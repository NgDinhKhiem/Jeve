```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Read the CSV file
df = pd.read_csv('memory_usage.csv')
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
# Plot HeapUsage over time
plt.figure(figsize=(12, 6))
plt.plot(df['ElapsedMs'], df['HeapUsage'], label='Heap Usage (bytes)')
plt.xlabel('Elapsed Time (ms)')
plt.ylabel('Heap Usage (bytes)')
plt.title('Heap Usage Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('heap_usage_over_time.png')
plt.show()
```

```python
# Plot ObjectCount over time
df = pd.read_csv('memory_usage.csv')
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
plt.figure(figsize=(10, 5))
plt.plot(df['ElapsedMs'], df['ObjectCount'], label='Object Count', color='orange')
plt.xlabel('Elapsed Time (ms)')
plt.ylabel('Object Count')
plt.title('Object Count Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('object_count_over_time.png')
plt.show()
```

```python
# Optionally, plot both on the same graph with twin axes
df = pd.read_csv('memory_usage.csv')
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
fig, ax1 = plt.subplots(figsize=(12, 6))
color = 'tab:blue'
ax1.set_xlabel('Elapsed Time (ms)')
ax1.set_ylabel('Heap Usage (bytes)', color=color)
ax1.plot(df['ElapsedMs'], df['HeapUsage'], color=color, label='Heap Usage')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Object Count', color=color)
ax2.plot(df['ElapsedMs'], df['ObjectCount'], color=color, label='Object Count')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Heap Usage and Object Count Over Time')
fig.tight_layout()
plt.savefig('heap_and_object_count_over_time.png')
plt.show()
```
