```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('./build/memory_usage.csv')

# Plot HeapUsage over ProcessCount
plt.figure(figsize=(12, 6))
plt.plot(df['ProcessCount'], df['HeapUsage'], label='Heap Usage (bytes)', color='blue')
plt.xlabel('Process Step')
plt.ylabel('Heap Usage (bytes)')
plt.title('Heap Usage Over Process Steps')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('heap_usage_over_steps.png')
plt.show()
```

```python
# Load data
df = pd.read_csv('./build/memory_usage.csv')

# Plot ObjectCount over ProcessCount
plt.figure(figsize=(10, 5))
plt.plot(df['TotalAllocations'], df['TotalAllocations'], label='Total Allocations Count', color='orange')
plt.xlabel('Process Step')
plt.ylabel('Object Count')
plt.title('Object Count Over Process Steps')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('object_count_over_steps.png')
plt.show()
```

```python

# Plot ObjectCount over ProcessCount
plt.figure(figsize=(10, 5))
plt.plot(df['ProcessCount'], df['ObjectCount'], label='Object Count', color='orange')
plt.xlabel('Process Step')
plt.ylabel('Object Count')
plt.title('Object Count Over Process Steps')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('object_count_over_steps.png')
plt.show()
```

```python
# Plot HeapUsage and ObjectCount with twin axes over ProcessCount
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:blue'
ax1.set_xlabel('Process Step')
ax1.set_ylabel('Heap Usage (bytes)', color=color)
ax1.plot(df['ProcessCount'], df['HeapUsage'], color=color, label='Heap Usage')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Object Count', color=color)
ax2.plot(df['ProcessCount'], df['ObjectCount'], color=color, label='Object Count')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Heap Usage and Object Count Over Process Steps')
fig.tight_layout()
plt.savefig('heap_and_object_count_over_steps.png')
plt.show()
```
